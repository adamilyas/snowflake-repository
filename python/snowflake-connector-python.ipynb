```python
# !pip3 install "snowflake-connector-python[pandas]"
import snowflake.connector
import getpass
```

```python
USER = input("User :: ")
PASSWORD = getpass.getpass("Password ::")
ACCOUNT_NAME = "123456.southeast-asia.azure" # CHANGE ME

# Data source
DATABASE = "SNOWFLAKE_SAMPLE_DATA"
SCHEMA = "tpcds_sf100tcl"

def connect():
    # Gets the version
    ctx = snowflake.connector.connect(
        user=USER,
        password=PASSWORD,
        account=ACCOUNT_NAME,
        database=DATABASE,
        schema=SCHEMA
        )
    
    return ctx

# Gets the version
ctx = connect()
cs = ctx.cursor()
try:
    cs.execute("SELECT current_version()")
    one_row = cs.fetchone()
    print(one_row[0])
finally:
    cs.close()
ctx.close()
```
